```python
import os
import polars as pl
import json
```

```python
m_dir_path = "C:/Users/aronn/OneDrive/Asztali g√©p/Rajk/Prog 2/c5_moves_with_valid_id"
moves_files = os.listdir(m_dir_path)
nunique_move_dict = {}
```

```python
for i, file in enumerate(moves_files):
    print(f"Processing file {i}")
    
    m_path = os.path.join(m_dir_path, file)

    moves = (
        pl.scan_parquet(m_path)
        .select(["moveNo", "move"])
        .group_by("moveNo")
        .agg(pl.col("move").n_unique().alias("nunique_moves"))
        .collect()
    )

    for row in moves.iter_rows():
        move_no = row[0]
        nunique_move = row[1]
        if move_no in nunique_move_dict:
            nunique_move_dict[move_no] += nunique_move
        else:
            nunique_move_dict[move_no] = nunique_move
```

Save to json (after sorting)

```python
sorted_dict = dict(sorted(nunique_move_dict.items(), key=lambda item: item[1], reverse=True))

with open("nunique_moves_per_move_number.json", "w") as json_file:
    json.dump(sorted_dict, json_file, indent=4)
```
