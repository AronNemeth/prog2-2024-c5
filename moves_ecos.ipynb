```python
import os
import polars as pl
```

Define variables

```python
m_dir_path = "C:/Users/aronn/OneDrive/Asztali gép/Rajk/Prog 2/c5_moves_with_valid_id"
moves_files = os.listdir(m_dir_path)
g_path = "C:/Users/aronn/OneDrive/Asztali gép/Rajk/Prog 2/games.parquet"
move_eco_dict = {}
```

Read in games

```python
games = (
    pl.scan_parquet(g_path)
    .select(["gameId", "ECO"])
    )
```

Calculate number of unique ECOs for each move

```python
for i, file in enumerate(moves_files):
    print(f"Processing file {i}")
    
    m_path = os.path.join(m_dir_path, file)

    # Number of unique ECOs for each move moves files
    moves_eco_n = (
        pl.scan_parquet(m_path)
        .select(["gameId", "moveNo", "move"])
        .filter(pl.col("moveNo") == "2.")
        .join(games, on="gameId", how="inner")
        .group_by("move")
        .agg(pl.col("ECO").n_unique().alias("nunique_ECOs"))
        .collect()
    )

    # Update dict
    for row in moves_eco_n.iter_rows():
        move_no = row[0]
        nunique_eco = row[1]
        if move_no in move_eco_dict:
            move_eco_dict[move_no] += nunique_eco
        else:
            move_eco_dict[move_no] = nunique_eco
```
